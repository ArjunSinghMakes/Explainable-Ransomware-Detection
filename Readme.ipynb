# 🛡️ Explainable Ransomware Detection using Machine Learning (XAI)

## 🚀 Overview

This project implements a Machine Learning–based system to detect ransomware attacks and enhance trust using **Explainable AI (XAI)** techniques.

Unlike traditional detection systems that only classify files as malicious or benign, this project also explains:

* **Why a file is classified as ransomware**
* **Which features contributed most to the decision**
* **What changes could alter the prediction (counterfactual reasoning)**



## 🎯 Problem Statement

Ransomware is one of the most critical cybersecurity threats. Traditional detection systems suffer from:

* Lack of transparency in predictions
* Difficulty in understanding model decisions
* Limited trust from security analysts

This project addresses these challenges by combining **Machine Learning + Explainability**.



## 💡 Solution

The system builds a complete pipeline:

* Data preprocessing
* Model training
* Performance evaluation
* Explainability using XAI methods
* Counterfactual analysis for actionable insights



## 🧠 Key Features

### 🔍 Ransomware Detection

* Classifies files/processes as:

  * **Ransomware**
  * **Benign**



### 🤖 Machine Learning Models

* Random Forest (primary model)
* Decision Tree
* Logistic Regression
* Naive Bayes



### 📊 Model Evaluation

* Accuracy
* Precision
* Recall
* F1-score

👉 Special focus on **Recall**, as missing ransomware is critical.



### 🧠 Explainable AI (XAI)

#### ✅ SHAP (SHapley Additive Explanations)

* Provides **global feature importance**
* Explains how each feature contributes to predictions



#### ✅ LIME (Local Interpretable Model-Agnostic Explanations)

* Explains **individual predictions**
* Shows feature contribution for a specific sample



### 🔥 Counterfactual Explanations (DiCE)

* Generates “what-if” scenarios
* Suggests minimal changes to alter predictions

👉 Example:

* “If API calls were reduced, the sample may not be classified as ransomware”
